```python
#apparently there's also yyyy-MM-dd HH:mm:SS and MM/dd/yyyy HH:mm

def loadData(loc):
    df = sqlContext.read.format("csv") \
    .option("header", "true") \
    .option("timestampFormat", "MM/dd/yyyy HH:mm:SS") \
    .option("inferSchema", "true") \
    .load(loc)
    
    return df
```

```python
def renameCols(df):
    df = df.withColumnRenamed("tripduration", "trip_duration")
    df = df.withColumnRenamed("starttime", "start_time")
    df = df.withColumnRenamed("stoptime", "stop_time")
    df = df.withColumnRenamed("start station id", "start_station_id")
    df = df.withColumnRenamed("start station name", "start_station_name")
    df = df.withColumnRenamed("start station latitude", "start_station_latitude")
    df = df.withColumnRenamed("start station longitude", "start_station_longitude")
    df = df.withColumnRenamed("end station id", "end_station_id")
    df = df.withColumnRenamed("end station name", "end_station_name")
    df = df.withColumnRenamed("end station latitude", "end_station_latitude")
    df = df.withColumnRenamed("end station longitude", "end_station_longitude")
    df = df.withColumnRenamed("birth year", "birth_year")
    
    return df

def renameWeirdCols(df):
    df = df.withColumnRenamed("Trip Duration", "trip_duration")
    df = df.withColumnRenamed("Start Time", "start_time")
    df = df.withColumnRenamed("Stop Time", "stop_time")
    df = df.withColumnRenamed("Start Station ID", "start_station_id")
    df = df.withColumnRenamed("Start Station Name", "start_station_name")
    df = df.withColumnRenamed("Start Station Latitude", "start_station_latitude")
    df = df.withColumnRenamed("Start Station Longitude", "start_station_longitude")
    df = df.withColumnRenamed("End Station ID", "end_station_id")
    df = df.withColumnRenamed("End Station Name", "end_station_name")
    df = df.withColumnRenamed("End Station Latitude", "end_station_latitude")
    df = df.withColumnRenamed("End Station Longitude", "end_station_longitude")
    df = df.withColumnRenamed("Birth Year", "birth_year")
    df = df.withColumnRenamed("User Type", "usertype")
    df = df.withColumnRenamed("Bike ID", "bikeid")
    df = df.withColumnRenamed("Gender", "gender")
    
    return df

def findRides(df):
    df = df.select(["trip_duration", "start_time", "stoptime", "start_station_id", "end_station_id", "bikeid"])
    return df

def findStations(df):
    df = df.select(['start_station_id', 'start_station_name', 'start_station_latitude', 'start_station_longitude', 'end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude'])
    return df
    
def cleanData(df):
    #remove nulls from trip_duration, start_time, stop_time, start_station_id, end_station_id, bikeid
    df = df.filter(df.trip_duration.isNotNull()).filter(df.trip_duration > 0)
    df = df.filter(df.start_time.isNotNull())
    df = df.filter(df.stop_time.isNotNull())
    df = df.filter(df.start_station_id.isNotNull())
    df = df.filter(df.end_station_id.isNotNull())
    df = df.filter(df.bikeid.isNotNull())
    return df
```

```python
def writeOutput(df, loc):
    df.repartition(1).write.option("header", "true").csv(loc)
```

```python
def aggregateFrames(df):
    #aggregates by hour, day, start and end
    df = df.withColumn('hour', hour(df.start_time)).withColumn('date', to_date(df.start_time))
    df = df.groupBy('date','hour','start_station_id').count()
    return df
    
def findStations(df):
    starts = df.select('start_station_id', 'start_station_name', 'start_station_latitude', 'start_station_longitude')
    starts = starts.withColumnRenamed('start_station_id', 'station_id') \
    .withColumnRenamed('start_station_name', 'station_name') \
    .withColumnRenamed('start_station_latitude', 'station_latitude') \
    .withColumnRenamed('start_station_longitude', 'station_longitude')

    ends = df.select('end_station_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude')
    ends = ends.withColumnRenamed('end_station_id', 'station_id') \
    .withColumnRenamed('end_station_name', 'station_name') \
    .withColumnRenamed('end_station_latitude', 'station_latitude') \
    .withColumnRenamed('end_station_longitude', 'station_longitude')

    stations = starts #.union(ends)
    stations = stations.distinct()
    return stations
```
